```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
si_data = pd.read_csv("run15_ivar16/SinkParticles.csv")
```

```python
active_sinks = np.where(si_data["Status"] != 0)[0]
```

```python
fig = plt.figure(figsize=(8, 6))


large = 18
medium = 15
small = 12

ax = fig.add_subplot(111)
mpluto = 0.01303*1e27 
ax.scatter(si_data["Mass"][active_sinks]/mpluto, 
           si_data["NumIce"][active_sinks] / (si_data["NumIce"][active_sinks] + si_data["NumSil"][active_sinks]),
           color='black')
ax.set_ylim(0.6, 1)
ax.set_xlim(9e-5,4e-3)

ax.set_xscale('log')

ax.set_ylabel('Ice Fraction',fontsize=large)
ax.set_xlabel(r'$M \, (M_{\rm Pluto})$',fontsize=large)

ax.tick_params(labelsize = large)

#ax.grid()

plt.savefig("mass_v_ice-fraction.pdf")
```
