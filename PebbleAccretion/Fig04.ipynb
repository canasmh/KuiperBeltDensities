```python
import numpy as np 

R=np.linspace(0,2000,100)
phi0=0.5
R2=1500.
R1=R2/10**phi0
print(R1)

tht = np.heaviside(R1-R,0.5)
box = np.heaviside(R-R1,0.5)-np.heaviside(R-R2,0.5)

phi = phi0*tht + box*np.log10(R2/R)

plot(R,phi,color='black')
xlabel(r"$R$ (km)")
ylabel(r"$\phi(R)$")
xlim([0,2000])
title("Porosity Function")
grid()
savefig("porosity_function.pdf")
```
