# Figure 8

```python
import numpy as np
import matplotlib.pyplot as plt
from cProfile import label
from matplotlib import colors
m_pluto = 1.309e25
import pandas as pd
from reader import KuiperBeltData
```

```python
kbos = KuiperBeltData()
data_dec = np.load("R_20_dec_rhopsmax_3_0_tmax_5827100_m_truncated_CO_Porosity.npz")
data_k = np.load("R_20_k_rhopsmax_3_0_tmax_5827800_m_truncated_CO.npz")
data_bi = np.load("R_20_bi_rhopsmax_3_2_tmax_5790000_m_truncated_CO.npz")
```

```python
t_dec = data_dec["t"] / 3600 / 24 / 365.25
dt_dec = data_dec["dt"]
mass_dec = data_dec["mass"]
porosity_dec = data_dec["porosity"]
density_dec = data_dec["density"]
ice_fraction_dec = data_dec["ice_fraction"]

n_dec = len(t_dec) - 1
it_dec = int(np.floor(n_dec / 3))
```

```python
t_k = data_k["t"] / 3600 / 24 / 365.25
dt_k = data_k["dt"]
mass_k = data_k["mass"]
porosity_k = data_k["porosity"]
density_k = data_k["density"]
ice_fraction_k = data_k["ice_fraction"]
n_k = len(t_k) - 1
it_k = int(np.floor(n_k / 3))
```

```python
t_bi = data_bi["t"] / 3600 / 24 / 365.25
dt_bi = data_bi["dt"]
mass_bi = data_bi["mass"]
porosity_bi = data_bi["porosity"]
density_bi = data_bi["density"]
ice_fraction_bi = data_bi["ice_fraction"]
n_bi = 309 - 1
it_bi = int(np.floor(n_bi / 3))
```

```python
fig = plt.figure(figsize=(20, 12))

cmap = plt.cm.seismic_r
norm = colors.BoundaryNorm(np.arange(0, 110, 10), cmap.N)

small = 10
medium = 14
large=16
rotation_angle = 90
    
ax1 = fig.add_subplot(3,4,1)
ax2 = fig.add_subplot(3,4,2, sharey=ax1)
ax3 = fig.add_subplot(3,4,3, sharey=ax1)
ax4 = fig.add_subplot(3,4,4, sharey=ax1)
ax5 = fig.add_subplot(3,4,5)
ax6 = fig.add_subplot(3,4,6, sharey=ax1)
ax7 = fig.add_subplot(3,4,7, sharey=ax1)
ax8 = fig.add_subplot(3,4,8, sharey=ax1)
ax9 = fig.add_subplot(3,4,9)
ax10 = fig.add_subplot(3,4,10, sharey=ax1)
ax11 = fig.add_subplot(3,4,11, sharey=ax1)
ax12 = fig.add_subplot(3,4,12, sharey=ax1)

ax1.scatter(mass_dec[it_dec * 0, :] / m_pluto, density_dec[it_dec * 0, :], c=ice_fraction_dec[it_dec * 0, :] * 100, cmap=cmap, norm=norm, s=9 ** 2, zorder=3)
ax1.set_xscale('log')
ax1.set_ylim(0, 3.3)
ax1.set_xlim(6e-5, 30)
ax1.set_ylabel(r'Density (g cm$^{-3}$)', fontsize=medium, labelpad=small)
ax1.grid(alpha=0.3)
ax1.tick_params(axis='x', labelsize=0, width=0)
ax1.tick_params(axis='y', labelsize=small, length=5)
ax1.text(2e-4, 2.8, f"T: {t_dec[it_dec * 0]:.3e} Years", fontsize=medium, backgroundcolor='#FFFFFF')

ax2.scatter(mass_dec[it_dec * 1, :] / m_pluto, density_dec[it_dec * 1, :], c=ice_fraction_dec[it_dec * 1, :] * 100, cmap=cmap, norm=norm, s=9 ** 2, zorder=3)
ax2.set_xscale('log')
ax2.set_xlim(6e-5, 30)
ax2.tick_params(axis='both', labelsize=0, width=0)
ax2.grid(alpha=0.3)
ax2.text(2e-4, 2.8, f"T: {t_dec[it_dec * 1]:.3e} Years", fontsize=medium, backgroundcolor='#FFFFFF')

ax3.scatter(mass_dec[it_dec * 2, :] / m_pluto, density_dec[it_dec * 2, :], c=ice_fraction_dec[it_dec * 2, :] * 100, cmap=cmap, norm=norm, s=9 ** 2, zorder=3)
ax3.set_xscale('log')
ax3.set_xlim(6e-5, 30)
ax3.tick_params(axis='both', labelsize=0, width=0)
ax3.grid(alpha=0.3)
ax3.text(2e-4, 2.8, f"T: {t_dec[it_dec * 2]:.3e} Years", fontsize=medium, backgroundcolor='#FFFFFF')


ax4.scatter(mass_dec[it_dec * 3, :] / m_pluto, density_dec[it_dec * 3, :], c=ice_fraction_dec[it_dec * 3, :] * 100, cmap=cmap, norm=norm, s=9 ** 2, zorder=3)
ax4.set_xscale('log')
ax4.set_xlim(6e-5, 30)
ax4.tick_params(axis='both', labelsize=0, width=0)
ax4.grid(alpha=0.3)
ax4.text(2e-4, 2.8, f"T: {t_dec[it_dec * 3]:.3e} Years", fontsize=medium, backgroundcolor='#FFFFFF')

ax5.scatter(mass_bi[it_bi * 0, :] / m_pluto, density_bi[it_bi * 0, :], c=ice_fraction_bi[it_bi * 0, :] * 100, cmap=cmap, norm=norm, s=9 ** 2, zorder=3)
ax5.set_xscale('log')
ax5.set_ylim(0, 3.3)
ax5.set_xlim(6e-5, 30)
ax5.set_ylabel(r'Density (g cm$^{-3}$)', fontsize=medium, labelpad=small)
ax5.tick_params(axis='x', labelsize=0, width=0)
ax5.tick_params(axis='y', labelsize=small, length=5)
ax5.grid(alpha=0.3)
ax5.text(2e-4, 2.8, f"T: {t_bi[it_bi * 0]:.3e} Years", fontsize=medium, backgroundcolor='#FFFFFF')

ax6.scatter(mass_bi[it_bi * 1, :] / m_pluto, density_bi[it_bi * 1, :], c=ice_fraction_bi[it_bi * 1, :] * 100, cmap=cmap, norm=norm, s=9 ** 2, zorder=3)
ax6.set_xscale('log')
ax6.set_xlim(6e-5, 30)
ax6.tick_params(axis='both', labelsize=0, width=0)
ax6.grid(alpha=0.3)
ax6.text(2e-4, 2.8, f"T: {t_bi[it_bi * 1]:.3e} Years", fontsize=medium, backgroundcolor='#FFFFFF')

ax7.scatter(mass_bi[it_bi * 2, :] / m_pluto, density_bi[it_bi * 2, :], c=ice_fraction_bi[it_bi * 2, :] * 100, cmap=cmap, norm=norm, s=9 ** 2, zorder=3)
ax7.set_xscale('log')
ax7.set_xlim(6e-5, 30)
ax7.tick_params(axis='both', labelsize=0, width=0)
ax7.grid(alpha=0.3)
ax7.text(2e-4, 2.8, f"T: {t_bi[it_bi * 2]:.3e} Years", fontsize=medium, backgroundcolor='#FFFFFF')

ax8.scatter(mass_bi[it_bi * 3, :] / m_pluto, density_bi[it_bi * 3, :], c=ice_fraction_bi[it_bi * 3, :] * 100, cmap=cmap, norm=norm, s=9 ** 2, zorder=3)
ax8.set_xscale('log')
ax8.set_xlim(6e-5, 30)
ax8.tick_params(axis='both', labelsize=0, width=0)
ax8.grid(alpha=0.3)
ax8.text(2e-4, 2.8, f"T: {t_bi[it_bi * 3]:.3e} Years", fontsize=medium, backgroundcolor='#FFFFFF')

ax9.scatter(mass_k[0 * it_k, :] / m_pluto, density_k[0 * it_k, :], c=ice_fraction_k[0 * it_k, :] * 100, cmap=cmap, norm=norm, s=9 ** 2, zorder=3)
ax9.set_xscale('log')
ax9.set_ylim(0, 3)
ax9.set_xlim(6e-5, 30)
ax9.set_ylabel(r'Density (g cm$^{-3}$)', fontsize=medium, labelpad=small)
ax9.set_xlabel(r'Mass (M$_{\rm{Pluto}}$)', fontsize=medium, labelpad=small)
ax9.grid(alpha=0.3)
ax9.text(2e-4, 2.8, f"T: {t_k[it_k * 0]:.3e} Years", fontsize=medium, backgroundcolor='#FFFFFF')
ax9.tick_params(axis='both', labelsize=small, length=5)

ax10.scatter(mass_k[1 * it_k, :] / m_pluto, density_k[1 * it_k, :], c=ice_fraction_k[1 * it_k, :] * 100, cmap=cmap, norm=norm, s=9 ** 2, zorder=3)
ax10.set_xscale('log')
ax10.set_xlim(6e-5, 30)
ax10.tick_params(axis='y', labelsize=0, width=0)
ax10.tick_params(axis='x', labelsize=small, length=5)
ax10.set_xlabel(r'Mass (M$_{\rm{Pluto}}$)', fontsize=medium, labelpad=small)
ax10.grid(alpha=0.3)
ax10.text(2e-4, 2.8, f"T: {t_k[it_k * 1]:.3e} Years", fontsize=medium, backgroundcolor='#FFFFFF')

ax11.scatter(mass_k[2 * it_k, :] / m_pluto, density_k[2 * it_k, :], c=ice_fraction_k[2 * it_k, :] * 100, cmap=cmap, norm=norm, s=9 ** 2, zorder=3)
ax11.set_xscale('log')
ax11.set_xlim(6e-5, 30)
ax11.tick_params(axis='y', labelsize=0, width=0)
ax11.tick_params(axis='x', labelsize=small, length=5)
ax11.set_xlabel(r'Mass (M$_{\rm{Pluto}}$)', fontsize=medium, labelpad=small)
ax11.grid(alpha=0.3)
ax11.text(2e-4, 2.8, f"T: {t_k[it_k * 2]:.3e} Years", fontsize=medium, backgroundcolor='#FFFFFF')

im = ax12.scatter(mass_k[3 * it_k, :] / m_pluto, density_k[3 * it_k, :], c=ice_fraction_k[3 * it_k, :] * 100, cmap=cmap, norm=norm, s=9 ** 2, zorder=3)
ax12.set_xscale('log')
ax12.set_xlim(6e-5, 30)
ax12.tick_params(axis='y', labelsize=0, width=0)
ax12.tick_params(axis='x', labelsize=small, length=5)
ax12.set_xlabel(r'Mass (M$_{\rm{Pluto}}$)', fontsize=medium, labelpad=small)
ax12.grid(alpha=0.3)
ax12.text(2e-4, 2.8, f"T: {t_k[it_k * 3]:.3e} Years", fontsize=medium, backgroundcolor='#FFFFFF')

plt.subplots_adjust(wspace=0, hspace=0)

axes = [ax1, ax2, ax3, ax4, ax5, ax6, ax7, ax8, ax9, ax10, ax11, ax12]

for ax in axes:
    
    ax.scatter(kbos.mass / m_pluto, kbos.density, marker="*", s=15 ** 2, c="darkgreen", zorder=2.5)
    ax.errorbar(x=kbos.mass / m_pluto, y=kbos.density, yerr=[kbos.min_density, kbos.max_density], ls='none', ecolor='darkgreen', zorder=2.5)
    ax.set_xscale('log')
    ax.grid(alpha=0.3)
    ax.set_xlim(6e-5, 30)
    ax.set_ylim(0, 3.3)

cbar_ax = fig.add_axes([0.815, 0.15, 0.01, 0.7])
fig.subplots_adjust(right=0.8)
cbar = fig.colorbar(im, cax=cbar_ax)

cbar_ax.tick_params(axis='y', labelsize=small, length=5)
cbar.set_label("Ice Fraction (%)", fontsize=medium, labelpad=medium, rotation=270)

label_x = -0.35
axes[0].text(label_x, 0.5, "Model 1 (Power Law)", fontsize=large, fontweight=600, rotation=rotation_angle,
                     ha='center', va='center', transform=axes[0].transAxes)

axes[4].text(label_x, 0.5, "Model 2 (Bimodal)", fontsize=large, fontweight=600, rotation=rotation_angle,
                     ha='center', va='center', transform=axes[4].transAxes)

axes[8].text(label_x, 0.5, "Model 3 (Constant)", fontsize=large, fontweight=600, rotation=rotation_angle,
                     ha='center', va='center', transform=axes[8].transAxes)

plt.savefig(fname="KBOs_v_t.pdf", format="pdf", dpi=350, bbox_inches='tight')
```
